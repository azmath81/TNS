```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Global YouTube Statistics.csv",encoding='latin1')

top_subscribers = df.sort_values(by='subscribers', ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.bar(top_subscribers['Youtuber'], top_subscribers['subscribers'])

plt.title("Top 10 YouTube Channels by Subscribers")
plt.xlabel("YouTube Channels")
plt.ylabel("Number of Subscribers")
plt.xticks(rotation=45)

plt.show()
print(top_subscribers)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Global YouTube Statistics.csv",encoding='latin1')

top_views = df.sort_values(by='video views', ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.bar(top_views['Youtuber'], top_views['video views'])

plt.title("Top 10 YouTube Channels by Video Views")
plt.xlabel("YouTube Channels")
plt.ylabel("Video Views")
plt.xticks(rotation=45)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Global YouTube Statistics.csv", encoding='latin1')

top_views = df.sort_values('video views', ascending=False).head(15)

plt.figure(figsize=(8,8))
plt.pie(top_views['video views'],
        labels=top_views['Youtuber'],
        autopct='%1.1f%%')
plt.title("Top 5 YouTube Channels by Video Views")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Global YouTube Statistics.csv",encoding='latin1')

top_countries = df['Country'].value_counts().head(10)

plt.figure(figsize=(8,8))
plt.pie(top_countries,
        labels=top_countries.index,
        autopct='%1.1f%%')

plt.title("Top 10 Countries by Number of YouTube Creators")
plt.show()
print(top_countries)
```
